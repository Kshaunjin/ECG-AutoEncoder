```python
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import os
import shutil

import wfdb
from wfdb import processing
```

讀取Normal type訊號

```python
#https://physionet.org/physiobank/database/challenge/2017/training/A00/
record17 = wfdb.rdrecord('A00001', pb_dir='challenge/2017/training/A00/')
wfdb.plot_wfdb(record=record17, title='Record A00001 from Physionet Challenge 2017')
display(record17.__dict__)
```

讀取Noise type訊號

```python
#https://physionet.org/physiobank/database/challenge/2017/training/A00/
record22 = wfdb.rdrecord('A00022', pb_dir='challenge/2017/training/A00/')
wfdb.plot_wfdb(record=record22, title='Record A00022 from Physionet Challenge 2017')
display(record22.__dict__)
```

查看訊號詳細資訊

```python
signals2, fields2 = wfdb.rdsamp('A00001', pb_dir='challenge/2017/training/A00/')
display(signals2)
display(fields2)
```

將訊號儲存為csv格式

```python
a = np.asarray(signals2)
np.savetxt("A00001.csv", a, delimiter=",")
```
